## Import Libraries & Load Dataset

```python
import numpy as np
import cv2
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense , Dropout , Flatten
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.utils import plot_model
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns
```

## Sheikh Dataset

```python
image_generator = ImageDataGenerator(
    rescale = 1.0 / 255.0,
    horizontal_flip = True,
    zoom_range = 0.2,
    validation_split = 0.2
)
```

```python
path_train = "/content/drive/MyDrive/Dataset/Sheykh-Recognition"
Train = image_generator.flow_from_directory(
    path_train,
    class_mode = "categorical",
    target_size = (224,224),
    batch_size = 32,
    subset = "training"
)
Val = image_generator.flow_from_directory(
    path_train,
    class_mode = "categorical",
    target_size = (224,224),
    batch_size = 32,
    subset = "validation",
)
```

# Model
# Transfer Learning

### MobileNetV2

```python
mobilenet_model = MobileNetV2(weights="imagenet",input_shape=(224,224,3),include_top=False)
```

```python
mobilenet_model.trainable = False
model = Sequential()
model.add(mobilenet_model)
model.add(Flatten())
model.add(Dense(512,activation="relu"))
model.add(Dense(256,activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(2,activation="softmax"))
```

```python
model.summary()
```

```python
plot_model(model,show_shapes=True)
```

```python
model.compile(optimizer=Adam(),loss="categorical_crossentropy",metrics=["accuracy"])
checkpoint = ModelCheckpoint("sheykh_model.h5",monitor="accuracy",mode="max",save_best_only=True)
```

```python
history = model.fit(
    Train,
    epochs = 20,
    validation_data = Val,
    callbacks = [checkpoint]
)
```

```python
fig , (ax1,ax2) = plt.subplots(1,2,figsize=(18,5))
ax1.set_title("Loss")
ax2.set_title("Accuracy")
ax1.plot(history.history["loss"],label="Train",marker=".")
ax1.plot(history.history["val_loss"],label="Test",marker=".")
ax2.plot(history.history["accuracy"],label="Train",marker=".")
ax2.plot(history.history["val_accuracy"],label="Test",marker=".")
ax1.legend()
ax2.legend()
plt.show()
```

```python
pred = []
for i in Val.filepaths:
  img = cv2.imread(i)
  img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
  img = cv2.resize(img,(224,224))
  img = img / 255.0
  img = img.reshape(1,224,224,3)
  pred.append(np.argmax(model.predict(img)))
pred = np.array(pred)
print(metrics.classification_report(Val.classes,pred))
```

```python
cm = metrics.confusion_matrix(Val.classes,pred)
plt.figure(figsize=(5,5))
sns.heatmap(cm,cmap ="YlGnBu",annot=True)
```
